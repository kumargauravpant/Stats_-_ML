```python
import numpy as np

np.random.seed(10)

# 100 random integers between 0 and 50
x = np.random.randint(0, 50, 100)

# Positive correlation with some noise
y = x + np.random.normal(0, 10, 100)

np.corrcoef(x, y) # Returns Pearson product-moment correlation coefficients
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.style.use('ggplot')

plt.scatter(x, y)
plt.show()
```

```python
# 100 random variable between 0 and 50
x = np.random.randint(0, 50, 100)

# Negative correlation with some noise
y = 100 - x + np.random.normal(0, 10, 100)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(0, 50, 100)
y = np.random.randint(0, 50, 100)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({'a': np.random.randint(0, 50, 100)})
df['b'] = df['a'] + np.random.normal(0, 10, 100) # Positively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.normal(0, 5, 100) # Negatively correlated with 'a'
df['d'] = np.random.randint(0, 50, 100) # not correlated with 'a'

df.corr() # Pearson correlation
```

```python
pd.plotting.scatter_matrix(df, figsize=(10, 10))
plt.show()
```
